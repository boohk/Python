```python
root = 'C:\\Users\\User\\Desktop\\현경_파이썬_코드\\Python\\'
stop_words_list = ['stopWords.txt', 'stopWords2.txt']
```

stopWords.txt는 쉼표(,)로 구분되며 'stopWords2.txt는 줄바꿈(\n)으로 구분되므로 split()에 각각에 맞는 구분자를 입력해야 한다.

```python
with open(str(root + stop_words_list[0]), encoding="utf-8",  errors='ignore') as f:
    sw_list1 = []
    for line in f:
        data = line.strip().split(',')
        sw_list1 += data
    print(sw_list1)
```

```python
# 중복 제거 
sw_list1 = list(set(sw_list1))
# print(sw_list1)
```

```python
with open(str(root + stop_words_list[1]), encoding="utf-8") as f:
    sw_list2 = []
    for line in f:
        data = line.strip().split('\n')
        sw_list2 += data
    print(sw_list2)
```

```python
total_stop_words = []
for sw in sw_list1:
    if sw == '':
        pass
    else:
        total_stop_words.append(sw)
# print(total_stop_words)        
```

```python
for sw in sw_list2:
    print(sw)
    total_stop_words.append(sw)
# print(total_stop_words)    
```

```python
# 혹시 모르니 한번 더 중복 제거 
total_stop_words = list(set(total_stop_words))
# print(total_stop_words)
```

```python
# 스탑리스트 저장(.txt)
saveRoot = 'C:\\Users\\User\\Desktop\\현경_파이썬_코드\\Python\\'
saveStopWords = open(str(root+'new_stopwords.txt'), 'w', encoding='utf-8')
for sw in total_stop_words:
    lastNum = int(len(total_stop_words)-1)
    if sw == total_stop_words[lastNum]:
         saveStopWords.write(sw)
    else:
        saveStopWords.write(sw+',')
saveStopWords.close()
```

```python
# 저장한 스탑리스트 읽기
with open(str(root+'new_stopwords.txt'), encoding="utf-8",  errors='ignore') as f:
    sw_list = []
    for line in f:
        data = line.strip().split(',')
        sw_list += data
    print(sw_list)
```

만들어진 스탑리스트이며 텍스트 파일인 스탑리스트를 호출하여 list형으로 변환 후 list에 내장 함수인 append를 사용하여 스탑리스트를 갱신할 수 있다.
- ex. total_stop_words.append('루팡')

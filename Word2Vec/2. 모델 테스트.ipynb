```python
from gensim.models import Word2Vec
import os
model_path_list = []
for (path, dir, files) in os.walk("D:\\Word2Vec_model_20180612"):
    for filename in files:
        ext = os.path.splitext(filename)[-1]
        if ext == '.model':
            model_path_list.append(str("%s\\%s" % (path, filename)))

model_list = []
for m in model_path_list:
    model_list.append(Word2Vec.load(m))
# print(model_list)
```

```python
cnt = 0
for m in model_list:
    cnt += 1
    print(cnt, m)
    print(cnt, m.most_similar(positive=['공포', '로맨스'], negative=['재미있다'], topn=2))
    print(cnt, m.most_similar('공포', topn=2))
```
